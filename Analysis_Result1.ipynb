# Adding libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Reading CSV file

```python
ac_df= pd.read_csv(r'AC_Data.csv')
```

```python
ac_df.head(5)
```

# Checking columns

```python
ac_df.columns
```

# Renaming Column 0 to Date

```python
ac_df=ac_df.rename(columns={'0': 'Date'})
```

```python
ac_df.shape
```

```python
ac_df.info()
```

# Checking Null Counts

```python
ac_df.isnull().sum()
```

# Converting all Nan with 0 or we can replace by use other techniques like replace by Mean/Avg of that columnm or interpolate the missing values

```python
ac_df=ac_df.replace(np.nan, 0)
```

```python
ac_df.head(5)
```

```python
ac_df.info()
```

```python
ac_df.isnull().sum()
```

```python
#getting only data from datatime column and Assemble that column in dataframe
ac_df['Date_Use'] =   pd.to_datetime(ac_df['Date']).dt.date
ac_df['Date_Use'] 
```

```python
ac_df.head(5)
```

```python
ac_df.info()
```

# Changing object type to date time

```python
ac_df[["Date", "Date_Use"]] = ac_df[["Date", "Date_Use"]].apply(pd.to_datetime)
```

```python
# ac_df['year'] = pd.DatetimeIndex(ac_df['Date_Use']).year
# ac_df['month'] = pd.DatetimeIndex(ac_df['Date_Use']).month
```

```python
ac_df['year'] = ac_df['Date_Use'].dt.year
ac_df['month'] =ac_df['Date_Use'].dt.month
```

# To check the Year and Month of data

```python
ac_df['year'].value_counts()
```

```python
ac_df.info()
```

```python
ac_df.head(5)
```

```python
ac_df=ac_df.drop(['year','month'], axis=1)
```

# Aggregating by 2 min Frequency but that doesnot make any sense

```python
ac_dfq=ac_df.groupby(pd.Grouper(key='Date', freq='2min')).sum()
ac_dfq.head(2)
```

```python
ac_dfq.shape
```

```python
ac_df.columns
```

```python
ac_df=ac_df.drop(['Date'], axis=1)
```

```python
ac_df.columns
```

# Aggregating day wise 

```python
ac_df_use_date=ac_df.groupby(pd.Grouper(key='Date_Use')).sum()
ac_df_use_date=ac_df_use_date.reset_index()
ac_df_use_date
```

```python
ac_df_use_date.columns
```

```python
ac_df_use=ac_df_use_date.drop(['Date_Use'], axis=1)
```

# Checking by Mean (Average)

```python
ac_df_use_mean=ac_df_use.mean(axis = 0)
ac_df_use_mean=pd.DataFrame(ac_df_use_mean)
ac_df_use_mean.T
```

```python
ac_df_use_mean.max()
```

```python
ac_df_use_mean.min()
```

```python
ac_df_use_mean.plot()   # Below we have Big Graph
```

# By Means/avg It seems like 'AC 18' having Highest Utilization and 'AC 5' having Least Utilization

# Checking by Total Sum of all two months data

```python
ac_df_use_max=ac_df_use.sum(axis = 0)
ac_df_use_max=pd.DataFrame(ac_df_use_max)
ac_df_use_max.T
```

```python
ac_df_use_max.max()
```

```python
ac_df_use_max.min()
```

```python
ac_df_use_max.plot()
```

# By Total SUM also It seems like 'AC 18' having Highest Utilization and  Least Utilization is 'AC 5'

# Final Answer
'AC 18' having Highest Utilization 


'AC 5' having Least Utilization
